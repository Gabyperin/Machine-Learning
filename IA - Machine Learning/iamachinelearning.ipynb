# Projeto 
### Prever o valor futuro de uma ação.

```python
import yfinance as yf
```

```python
codigo = input("Digite o codigo de ação da empresa: ")
dados = yf.Ticker(codigo).history("2y")
dados.head()
```

## Tratamento de Dados

```python
# resetar o indice 
treinamento = dados.reset_index()

# extrair so a data da coluna Date
treinamento["Date"] = treinamento["Date"].dt.date
treinamento

# estraindo apenas a coluna Date e Close
colunas = ["Date", "Close"]
treinamento = treinamento[colunas]

# alterar o nome da coluna
treinamento.columns = ["ds", "y"]
treinamento

```

## Criando e Treinando o modelo Machine Learning 

```python
!pip install prophet
```

```python
from prophet import Prophet
```

```python
# criando o modelo
modelo = Prophet()
```

```python
# treinando nossa machine learning 
modelo.fit(treinamento)
```

## Gerando Previsões

```python
# quantos dias a mais quer que gere as ações 
periodo= modelo.make_future_dataframe(90)
periodo
```

```python
previsoes = modelo.predict(periodo)
previsoes
```

## Visualizando os dados 

```python
from prophet.plot import plot_plotly
```

```python
plot_plotly(modelo, previsoes)
```
